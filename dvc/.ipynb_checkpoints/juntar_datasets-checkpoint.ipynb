```python
def obtain_data(data):
    data_d = json.loads(data)
    if len(data_d) == 1:
        return data_d[0]['name']
    return [x['name'] for x in data_d]
```

```python
cols_decode = ["genres", "keywords", "production_companies", "production_countries", "spoken_languages"]
```

```python
for c in cols_decode:
    data2[c] = data2[c].apply(obtain_data)
```
